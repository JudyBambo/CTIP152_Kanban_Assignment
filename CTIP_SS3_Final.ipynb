```python
# Variables to store user and task data
username = ""
password = ""
first_name = ""
last_name = ""
tasks = []
task_counter = 1
todo_tasks = []
doing_tasks = []
done_tasks = []



# Task management function
def task_management():
    while True:
        print("\n--- Task Management ---")
        print("1. Add Task\n2. Show Report\n3. Exit")
        
        try:
            option = int(input("Choose an option (1,2,3): "))

        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
                
        if option == 1:
            add_task()
        elif option == 2:
            task_report() 
        elif option == 3:
            print("---Exiting the Task Management System---")
            break
        else:
            print("Please select a valid option (1, 2, 3)")


# Function for adding tasks
def add_task():
    global task_counter, tasks, todo_tasks, doing_tasks, done_tasks
    try:
        num_tasks = int(input("How many tasks would you like to add? "))
    #while not num_tasks.isdigit():
    except ValueError:  
        print("Invalid input. Please enter a number.")
        num_tasks = int(input("How many tasks would you like to add? "))
    
    for i in range(num_tasks):
        print(f"\n--- Task {i+1} ---")
        task_name = input("Enter task name: ")
        developer_name = input("Enter the developer's name (First & Last): ")
        
        while True:
            try:
                task_duration = int(input("Enter the duration of this task (hours): "))
                break
            except ValueError:
                print("Invalid input")
                continue
        
        while True:
            task_status = input("What is the status of your task (To-do/Doing/Done)?:").capitalize()
            if task_status in ["To-do", "Doing", "Done"]:
                break
            else:
                print("Invalid status")
            
        while True:
            description = input("Enter task description (max 50 characters): ")
            if len(description) > 50:
                print("Description too long. Try again.")
            else:
                break

    #Creating a dictionary to store our tasks
        task = {
            "Task Number": f"TASK {task_counter}",
            "Task Name": task_name,
            "Task Description": description,
            "Task Status": task_status,
            "Developer Name": developer_name,
            "Task Duration": task_duration
        }

        #Catergorizing task_status 
        if task_status == "To-do":
            todo_tasks.append(task)

        elif task_status == "Doing":
            doing_tasks.append(task)

        elif task_status == "Done":
            done_tasks.append(task)
        
        tasks.append(task)
        task_counter += 1
        print("Task added successfully!")


#Function to display tasks
def display_tasks(task_list, title):
    print(f"----{title}----")
    if len(task_list) == 0:
        print("No tasks")
    else:
        for task in task_list:
            for key, value in task.items():
                print(f"{key}: {value}")
            print("\n")

#Task report function        
def task_report():
    global task_counter, tasks, todo_tasks, doing_tasks, done_tasks
   #The function for displaying tasks
                
    while True:
        print("\n--- Task Report ---")
        print("1. Display All Tasks")
        print("2. Display Completed Tasks (Done)")
        print("3. Display Longest Task")
        print("4. Search by Task Name")
        print("5. Search by Developer's Name")
        print("6. Display Total Tasks Duration")
        print("7. Delete Task")
        print("8. Exit Report")
        
        choice = int(input("Enter an option (1-8): "))

        if choice == 1:
           display_tasks(tasks, "All Tasks") 
        
        elif choice == 2:
            display_tasks(done_tasks, "Done Tasks")

        elif choice == 3:
            longest_task()
            
        elif choice == 4:
            search_task()
            
        elif choice == 5:
            search_developer()
            
        elif choice == 6:
            total_duration()

        elif choice == 7:
            delete_task()
            
        elif choice == 8:
            break
        else:
            print("Invalid choice! Enter a number between 1 and 8!")

def longest_task():
    try:
        longest_task = tasks[0]
        for task in tasks[1]:
            if task["Task Duration"] > longest_task["Task Duration"]:
                longest_task = task
        display_tasks([longest_task], "Longest Task")

    except IndexError:
        print("No tasks")


def search_task():    
    search_task = input("Search by entering Task name: ")
    found = False
    for task in tasks:   
        if search_task == task["Task Name"]:
            for key, value in task.items():
                print(f"{key}: {value}")
            found = True
            break
    if not found:
        print("No task in that name!")


def search_developer():
    search_name = input("Search by Developer's Fullname (First & last): ")
    found = False
    print(f"----Tasks by {search_name.title()}----")
    
    for task in tasks:
        if search_name == task["Developer Name"]:
           
            for key, value in task.items():
                print(f"{key}: {value}")
            found = True
            
    if not found:
        print("No task by this developer.")

            
# Function to calculate the total tasks duration
def total_duration():
    global tasks
    total_duration = 0
    for task in tasks:
        total_duration += task["Task Duration"]
    print(f"Total duration of all tasks:{total_duration} hours")
    
            
# function to delete tasks
def delete_task():
    global tasks, todo_tasks, doing_tasks, done_tasks
    task_name = input("Enter the name of the task to be deleted: ")
    found = False
    for task in tasks:
        if task["Task Name"] == task_name:
            tasks.remove(task)
            if task in todo_tasks:
                todo_tasks.remove(task)
            elif task in doing_tasks:
                doing_tasks.remove(task)
            elif task in done_tasks:
                done_tasks.remove(task)
            print(f"\n'{task_name}' deleted.")
            found = True
            break
    if not found:
        print("Task not found.")
    


# User login function
def user_login(username, password):
    print("--- User Login ---")
    while True:
        input_username = input("Enter your username: ")
        if input_username != username:
            print("Invalid username")
            continue
        input_password = input("Enter your password: ")
        if input_password != password:
            print("Invalid password")
            continue
        print("--- Successfully Logged in ---")
        print(f"--- Welcome {first_name} {last_name}! ---")
        task_management()
        break

# Username validation
def user_name_val():
    while True:
        user_name = input("Create a Username: ")
        if len(user_name) > 5 or "_" not in user_name:
            print("Invalid username: Must be less than 5 characters and contain '_'")
        else:
            print("Valid username")
            return user_name


# Password validation
def password_val():
    while True:
        password = input("Create a Password: ")
        if len(password) < 8:
            print("Invalid password: Password must have 8+ characters")
            continue

        capital = False
        number = False
        special_char = False

        for char in password:
            if char.isupper():
                capital = True
            if char.isdigit():
                number = True
            if not char.isalnum():
                special_char = True

        if not capital:
            print("Invalid, Password must contain at least 1 capital letter")
        if not number:
            print("Invalid, Password must contain at least 1 number")
        if not special_char:
            print("Invalid, Password must contain at least 1 special character")

        if capital and number and special_char:
            print("Valid Password")
            return password


# User registration function
def user_registration():
    global username, password, first_name, last_name
    print("--- User Registration ---")
    first_name = input("Enter your first name: ")
    last_name = input("Enter your last name: ")
    username = user_name_val()
    password = password_val()
    print("---- Successfully registered!----")
    user_login(username, password)



# Welcome page function
def welcome_page():
    print("--- Welcome to Basic KanBan ---")
    print("1. Register\n2. Login\n3. Exit")
    option = input("Select an option (1,2,3): ")

    while option not in ['1', '2', '3']:
        print("Invalid option. Please enter 1, 2, or 3.")
        welcome_page()

    return int(option)
    

# Main menu (Entry point)
while True:
    options = welcome_page()
    if options == 1:
        user_registration()
    elif options == 2:
        if username == "" or password == "":
            print("User not registered yet. Please register first")
        else:
            user_login(username, password)
    elif options == 3:
        print("--- Thank you for using Basic KanBan ---")
        break
```
